```python
df.drop(columns = ['Unnamed: 0'] , inplace = True)
df.head(3)
```

```python
def compute_profit_per_unit(x):
    x["Profit/Unit"] = (x["จำนวนเงินทั้งหมด"] - x["ต้นทุนขายหักคูปองและcoin"])/x["จำนวน"]

    return x

df = df.apply(compute_profit_per_unit , axis = 1)

df.head(3)
```
