```python
import pandas as pd
```

```python
df = pd.read_csv(r"data\filtered_shipping_prices_august.csv")

df.head(3)
```

```python
df2 = pd.read_csv(r"data\filtered_shipping_prices_july.csv")

full = pd.concat([df , df2] , axis = 0)

full = full[["จังหวัด" ,  'ค่าจัดส่งที่ชำระโดยผู้ซื้อ']]
full = full.reset_index(drop = True)
full
```

```python
df = df[["จังหวัด" ,  'ค่าจัดส่งที่ชำระโดยผู้ซื้อ']]

df.head(3)
```

```python
df.columns
```

```python
city_frequency = full["จังหวัด"].value_counts()

ratio = city_frequency/full.shape[0]

df["City frequency"] = df["จังหวัด"].map(ratio)

df
```
