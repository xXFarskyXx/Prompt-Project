```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("data\ต้นทุกนสินค้า buildmate ที่ขายใน shopee ปี 67.xlsx")

df.head(3)
```

```python
df.columns
```

```python
df.columns = df.iloc[0]

df
```

```python
df.drop(0 , inplace = True)

df
```

```python
df.columns
```

```python
df.drop(columns = df.columns[0] , inplace = True)
df.drop(columns = ["#" , "วันที่"] , inplace = True)
df = df.reset_index(drop = True)

df.head()
```

```python
df["ขาย (บาท)"] = df["ขาย (บาท)"]/df["จำนวน"]
df["ต้นทุน (บาท)"] = df["ต้นทุน (บาท)"]/df["จำนวน"]
df["กำไร (บาท)"] = df["กำไร (บาท)"]/df["จำนวน"]

df
```

```python
mean_cost = df.groupby('สินค้า')['ต้นทุน (บาท)'].mean()

mean_cost_df = mean_cost.reset_index()
mean_cost_df.columns = ['สินค้า', 'ค่าเฉลี่ยต้นทุน (บาท)']
# Display the result

mean_cost_df.rename(columns= {"สินค้า" : "Product Name" , "ค่าเฉลี่ยต้นทุน (บาท)" : "COGS/Unit"} , inplace = True)
mean_cost_df
```

```python
mean_cost_df.to_excel("COGS_Modify.xlsx")
```
